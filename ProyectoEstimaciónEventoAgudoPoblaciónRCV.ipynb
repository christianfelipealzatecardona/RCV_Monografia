### **Contexto del problema**


Enfermedad Cardiovascular Son enfermedades que afectan a las arterias del corazón y del resto del organismo, principalmente el cerebro, los riñones y los miembros inferiores. 

Las más importantes son: el infarto de miocardio y el accidente cerebrovascular (trombosis, embolia y hemorragia cerebral). Son muy graves y la principal causa de muerte sobre todo en países desarrollados. 


Riesgo Cardiovascular es la probabilidad que tiene un individuo de sufrir una de estas enfermedades dentro de un determinado plazo de tiempo. Depende de los factores de riesgo *(Edad, Sexo, Raza, Antecedentes Familiares, Hipertensión, Colesterol, Diabetes, Tabaquismo, Sedentarismo, Alcohol, Ansiedad, Estrés)*.

Se realizará una clasificación binaria para predecir con base a un conjunto de variables clinicas si un paciente en condición de riesgo cardiovascular es susceptible de presentar un evento agudo (una hospitalizacion o una urgencia).


Para ello se realizará un análisis exploratorio de las variables evaluando que variables aportan a la predicción, posteriormente se dividirán los datos en conjunto de entrenamiento (70%) y conjunto de validación (30%) y el entrenamiento se hará utilizando validación cruzada aleatoria.

Finalmente en esta primera iteracion el algoritmo de clasificación mostrará la exactitud sobre el conjunto de validación, para la ultima iteracion que su desempeño de clasificación sea superior al azar.

### **Estructura del documento**

Para resolver el problema, se utilizarán técnicas de Machine Learning que permitan separar a la población con base en si presentaron o no eventos agudos (hospitalización o urgencias).

Se utilizarán 3 modelos:

**Clasificación:** Con este método se puede separar las personas entre las que presentan eventos de agravamiento .
* Análisis discriminante cuadrático *(QuadraticDiscriminant)* 





**Predicción:** Permite estimar la probabilidad de presentar un evento de agravamiento.
* Regresión logística *(LogisticRegression)* 
* Clasificador bayesiano ingenuo *(GaussianNB)* 


**Modelos a considerar**


Para abordar este problema de clasificación se eligen inicialmente 3 algoritmos incluidos en el paquete de sklearn:

*Análisis discriminante cuadrático - (QuadraticDiscriminant)* Es un clasificador estadístico que tiene la ventaja de ser fácilmente computable, no tiene hiperparámetros para ajustar, que es inherentemente multiclase y ha probado trabajar bien en la práctica [1]

*Clasificador bayesiano ingenuo - (GaussianNB)* Es un algoritmo de aprendizaje supervisado que se basa en la aplicación del teorema de bayes, asumiendo la independencia condicional dentro de cada par de atributos. Su esimación es rápida, y dado que no asume dependencia, se evita problema derivados por la dimensionalidad. [2]

*Regresión logística - (LogisticRegression)* Es un algoritmo de predicción de resultado de variables categóricas, es decir, se puede utilizar para clasificación en variables binarias. Su salida es la probabilidad de las diferentes salidas de una variable de respuesta, computada utilizando una función logística [3]

[1] Scikit-learn.org. 2021. 1.2. Linear and Quadratic Discriminant Analysis — scikit-learn 0.24.2 documentation. [online] Available at: <https://scikit-learn.org/stable/modules/lda_qda.html#lda-qda> [Accessed 2 July 2021].

[2] Scikit-learn.org. 2021. 1.9. Naive Bayes — scikit-learn 0.24.2 documentation. [online] Available at: <https://scikit-learn.org/stable/modules/naive_bayes.html> [Accessed 2 July 2021].
