# **Machine Learning aplicado a la estimación de eventos agudos en población en condición de riesgo cardiovascular- RCV**

### **Contexto del problema**


Enfermedad Cardiovascular Son enfermedades que afectan a las arterias del corazón y del resto del organismo, principalmente el cerebro, los riñones y los miembros inferiores. 

Las más importantes son: el infarto de miocardio y el accidente cerebrovascular (trombosis, embolia y hemorragia cerebral). Son muy graves y la principal causa de muerte sobre todo en países desarrollados. 


Riesgo Cardiovascular es la probabilidad que tiene un individuo de sufrir una de estas enfermedades dentro de un determinado plazo de tiempo. Depende de los factores de riesgo *(Edad, Sexo, Raza, Antecedentes Familiares, Hipertensión, Colesterol, Diabetes, Tabaquismo, Sedentarismo, Alcohol, Ansiedad, Estrés)*.

Se realizará una clasificación binaria para predecir con base a un conjunto de variables clinicas si un paciente en condición de riesgo cardiovascular es susceptible de presentar un evento agudo (una hospitalizacion o una urgencia).


Para ello se realizará un análisis exploratorio de las variables evaluando que variables aportan a la predicción, posteriormente se dividirán los datos en conjunto de entrenamiento (70%) y conjunto de validación (30%) y el entrenamiento se hará utilizando validación cruzada aleatoria.

Finalmente en esta primera iteracion el algoritmo de clasificación mostrará la exactitud sobre el conjunto de validación, para la ultima iteracion que su desempeño de clasificación sea superior al azar.

### **Estructura del documento**

Para resolver el problema, se utilizarán técnicas de Machine Learning que permitan separar a la población con base en si presentaron o no eventos agudos (hospitalización o urgencias).

Se utilizarán 3 modelos:

**Clasificación:** Con este método se puede separar las personas entre las que presentan eventos de agravamiento .
* Análisis discriminante cuadrático *(QuadraticDiscriminant)* 





**Predicción:** Permite estimar la probabilidad de presentar un evento de agravamiento.
* Regresión logística *(LogisticRegression)* 
* Clasificador bayesiano ingenuo *(GaussianNB)* 


**Modelos a considerar**


Para abordar este problema de clasificación se eligen inicialmente 3 algoritmos incluidos en el paquete de sklearn:

*Análisis discriminante cuadrático - (QuadraticDiscriminant)* Es un clasificador estadístico que tiene la ventaja de ser fácilmente computable, no tiene hiperparámetros para ajustar, que es inherentemente multiclase y ha probado trabajar bien en la práctica [1]

*Clasificador bayesiano ingenuo - (GaussianNB)* Es un algoritmo de aprendizaje supervisado que se basa en la aplicación del teorema de bayes, asumiendo la independencia condicional dentro de cada par de atributos. Su esimación es rápida, y dado que no asume dependencia, se evita problema derivados por la dimensionalidad. [2]

*Regresión logística - (LogisticRegression)* Es un algoritmo de predicción de resultado de variables categóricas, es decir, se puede utilizar para clasificación en variables binarias. Su salida es la probabilidad de las diferentes salidas de una variable de respuesta, computada utilizando una función logística [3]

[1] Scikit-learn.org. 2021. 1.2. Linear and Quadratic Discriminant Analysis — scikit-learn 0.24.2 documentation. [online] Available at: <https://scikit-learn.org/stable/modules/lda_qda.html#lda-qda> [Accessed 2 July 2021].

[2] Scikit-learn.org. 2021. 1.9. Naive Bayes — scikit-learn 0.24.2 documentation. [online] Available at: <https://scikit-learn.org/stable/modules/naive_bayes.html> [Accessed 2 July 2021].

##### Análisis de datos atípicos

Se realiza en las variable numérica Edad, ya que son las que presentan mayor dispersión

#### 6. Modelamiento

Se abordará el problema utilizando 3 modelos diferentes:

**Clasificación:** Con este método se puede separar las personas entre las que presentan eventos de agravamiento .
* Análisis discriminante cuadrático *(QuadraticDiscriminant)* 





**Predicción:** Permite estimar la probabilidad de presentar un evento de agravamiento.
* Regresión logística *(LogisticRegression)* 
* Clasificador bayesiano ingenuo *(GaussianNB)* 


Por último se medirá el desempeño de cada uno de los modelos




```python
X = df_dummys.drop(['Y'],axis=1) 
y = df_dummys['Y']  
```

```python
# dividimos el conjunto de datos en conjunto de entrenamiento y conjunto de pruebas dado el desbalance de los datos se estratifica
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, stratify =y)
```

```python
# estandarizar variables
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

En la primera iteración de resolver el problema de presentar un evento agudo, con los parámetros por defecto para cada modelo y entrenarlos con el conjunto X_train, y_train.

modelos de regresión logística, Naïve Bayes y Discriminante Cuadrático para 

1.   Discriminante Cuadrático
2.   Naïve Bayes
3.   Regresión logística



```python
# 1. Modelo Discriminante Cuadrático

from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
clf1 = QuadraticDiscriminantAnalysis().fit(X_train,y_train)


from sklearn.metrics import classification_report
cr = classification_report(y_test, clf1.predict(X_test))

print('\n')
print('Reporte de clasificación Modelo Discriminante Cuadrático')
print('\n')
print(cr)

print('\n')

#Matriz de confusión para Discriminante Cuadrático
from sklearn.metrics import plot_confusion_matrix
disp = plot_confusion_matrix(clf1, X_test, y_test, cmap=plt.cm.Blues)
disp.ax_.set_title('Matriz de confusión - Discriminante Cuadrático');
```

```python
# print(roc_auc_score(ytest, p_train1))
```

```python
# 2. Modelo Naïve Bayes

from sklearn.naive_bayes import GaussianNB
clf2 = GaussianNB().fit(X_train,y_train)


from sklearn.metrics import classification_report
cr = classification_report(y_test, clf2.predict(X_test))

print('\n')
print('Reporte de clasificación GaussianNB')
print('\n')
print(cr)

print('\n')

#Matriz de confusión para Naïve Bayes
from sklearn.metrics import plot_confusion_matrix
disp = plot_confusion_matrix(clf2, X_test, y_test, cmap=plt.cm.Blues)
disp.ax_.set_title('Matriz de confusión - GaussianNB');
```

```python
# 3. Regresión Logística

from sklearn.linear_model import LogisticRegression
clf3 = LogisticRegression(random_state=0,class_weight='balanced').fit(X_train,y_train)


from sklearn.metrics import classification_report
cr = classification_report(y_test, clf3.predict(X_test))

print('\n')
print('Reporte de clasificación Regresión Logística')
print('\n')
print(cr)

print('\n')

#Matriz de confusión para Regresión Logística
from sklearn.metrics import plot_confusion_matrix
disp = plot_confusion_matrix(clf3, X_test, y_test, cmap=plt.cm.Blues)
disp.ax_.set_title('Matriz de confusión - Regresión Logística');
```

```python
#  Curva ROC, del modelo escogido
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
logit_roc_auc = roc_auc_score(y_test, clf3.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, clf3.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```

tenemos un AUC del 80% que nos permite hablar de un buen poder predictivo del modelo ante diferentes cutoff y en consecuencia elegir un punto de corte que maximice la sensibilidad y la especificidad del clasificador

se decide recalcular el cutoff de la regresion logistica con el objetivo de tener una sensibilidad y una especificidad balanceada, de tal manera que la precision del modelo a la hora de predecir tanto a un individuo con riesgo de presentar un evento agudo como a uno que no lo presente sea igual.

```python
from scipy.stats import norm

def logit_pvalue(model, x):
  """ Calculate z-scores for scikit-learn LogisticRegression.
  parameters:
  model: fitted sklearn.linear_model.LogisticRegression with intercept and large C
  x: matrix on which the model was fit
  This function uses asymtptics for maximum likelihood estimates.
    """
  p = model.predict_proba(x)
  n = len(p)
  m = len(model.coef_[0]) + 1
  coefs = np.concatenate([model.intercept_, model.coef_[0]])
  x_full = np.matrix(np.insert(np.array(x), 0, 1, axis = 1))
  ans = np.zeros((m, m))
  for i in range(n):
    ans = ans + np.dot(np.transpose(x_full[i, :]), x_full[i, :]) * p[i,1] * p[i, 0]
  vcov = np.linalg.inv(np.matrix(ans))
  se = np.sqrt(np.diag(vcov))
  t = coefs/se
  p = (1 - norm.cdf(abs(t))) * 2
  return p



p_value=logit_pvalue(clf3, X_train)
p_value
```
